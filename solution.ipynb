## DATA PREPROCESSING

* ONE HOT ENCODING ON CATEGORICAL VARIABLES
* STANDARD SCALING ON CONTINUOUS VARIABLES

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(categories="auto")
```

```python
data["SEX"]=encoder.fit_transform(data["SEX"].values.reshape(-1,1)).toarray()
```

```python
data["EDUCATION"]=encoder.fit_transform(data["EDUCATION"].values.reshape(-1,1)).toarray()
```

```python
data["MARRIAGE"]=encoder.fit_transform(data["MARRIAGE"].values.reshape(-1,1)).toarray()
data["PAY_0"]=encoder.fit_transform(data["PAY_0"].values.reshape(-1,1)).toarray()
data["PAY_2"]=encoder.fit_transform(data["PAY_2"].values.reshape(-1,1)).toarray()
data["PAY_3"]=encoder.fit_transform(data["PAY_3"].values.reshape(-1,1)).toarray()
data["PAY_4"]=encoder.fit_transform(data["PAY_4"].values.reshape(-1,1)).toarray()
data["PAY_5"]=encoder.fit_transform(data["PAY_5"].values.reshape(-1,1)).toarray()
data["PAY_6"]=encoder.fit_transform(data["PAY_6"].values.reshape(-1,1)).toarray()
```

## CHECK FOR MISSING VALUES

```python
if True in data.isna().values==True:
    print("Data has missing values")
else:
    print("No missing values")
# No missing values
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data['BILL_AMT1'] = scaler.fit_transform(data['BILL_AMT1'].values.reshape(-1,1))
data['BILL_AMT2'] = scaler.fit_transform(data['BILL_AMT2'].values.reshape(-1,1))
data['BILL_AMT3'] = scaler.fit_transform(data['BILL_AMT3'].values.reshape(-1,1))
data['BILL_AMT4'] = scaler.fit_transform(data['BILL_AMT4'].values.reshape(-1,1))
data['BILL_AMT5'] = scaler.fit_transform(data['BILL_AMT5'].values.reshape(-1,1))
data['BILL_AMT6'] = scaler.fit_transform(data['BILL_AMT6'].values.reshape(-1,1))
data['PAY_AMT1'] = scaler.fit_transform(data['PAY_AMT1'].values.reshape(-1,1))
data['PAY_AMT2'] = scaler.fit_transform(data['PAY_AMT2'].values.reshape(-1,1))
data['PAY_AMT3'] = scaler.fit_transform(data['PAY_AMT3'].values.reshape(-1,1))
data['PAY_AMT4'] = scaler.fit_transform(data['PAY_AMT4'].values.reshape(-1,1))
data['PAY_AMT5'] = scaler.fit_transform(data['PAY_AMT5'].values.reshape(-1,1))
data['PAY_AMT6'] = scaler.fit_transform(data['PAY_AMT6'].values.reshape(-1,1))
```

## MODEL TRAINING AND PREDICTION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

## REMOVING UNNECESSARY COLUMNS IN DATASET

```python
data.drop(["default payment next month"],inplace=True,axis=1)
data.drop(["ID"],inplace=True,axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.3,shuffle=True)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

## EVALUATING THE PERFORMANCE USING ACCURACY SCORE

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
